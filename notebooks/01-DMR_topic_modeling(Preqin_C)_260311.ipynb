## 1. DMR 학습

아래 예시는 문장 단위 데이터 기준이다. 문단 단위 분석은 `--token-col pg_token --text-col paragraph`로 바꾸어 실행한다.

```python
!python -m preqin_topic_modeling.train_dmr \
  --data-path ../data/processed/st_df_All_fin_251120_nega.pkl \
  --output-dir ../outputs \
  --model-name dmr_k100_min30_rm0.bin \
  --token-col token \
  --text-col sentence \
  --metadata-col category \
  --era-col Era \
  --era-value 先秦 \
  --k 100 \
  --min-df 30 \
  --rm-top 0 \
  --seed 2025 \
  --term-weight PMI \
  --iterations 500 \
  --step 20 \
  --top-n-words 30
```
